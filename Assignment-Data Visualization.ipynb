```python
from google.colab import drive
import pandas as pd

# Mount Google Drive
drive.mount('/content/drive')

# Ubah path file sesuai dengan lokasi file dataset Anda di Google Drive
file_path = '/content/drive/My Drive/Colab Notebooks/epl-goalScorer(20-21).csv'

# Baca dataset dari file CSV
football_data = pd.read_csv(file_path)

# 10 Player terbanyak mencetak gol
top_10_scorers = football_data.sort_values(by='goals', ascending=False).head(10)

# 10 player terlama durasi bermain
top_10_duration = football_data.sort_values(by='time', ascending=False).head(10)

# 10 player terbanyak assist
top_10_assists = football_data.sort_values(by='assists', ascending=False).head(10)

# 4 player pencetak gol terbanyak dari kesebelasan Manchester City
goals_manchester_city = football_data[football_data['team_title'] == 'Manchester City'].sort_values(by='goals', ascending=False).head(4)

# Dari kesebelasan mana pencetak gol terbanyak
most_goals_team = football_data.groupby('team_title')['goals'].sum().idxmax()

# Jumlah kartu kuning dan merah yang diterima pembuat assist terbanyak
most_assist_player = football_data.loc[football_data['assists'].idxmax()]
yellow_cards_most_assist = most_assist_player['yellow_cards']
red_cards_most_assist = most_assist_player['red_cards']

# Jumlah assist dan golnya paling banyak dan dari kesebelasan mana
most_goals_assists_player = football_data.loc[(football_data['goals'] + football_data['assists']).idxmax()]
most_goals_assists_team = most_goals_assists_player['team_title']
most_goals_assists = most_goals_assists_player['goals'] + most_goals_assists_player['assists']

# 5 kesebelasan yang menyumbangkan gol terbanyak
top_5_goal_contributors = football_data.groupby('team_title')['goals'].sum().nlargest(5)

# Urutkan pemain berdasarkan jumlah gol terbanyak tanpa mendapatkan kartu sama sekali
players_without_cards = football_data[(football_data['yellow_cards'] == 0) & (football_data['red_cards'] == 0)].sort_values(by='goals', ascending=False)

print("10 Player terbanyak mencetak gol:")
print(top_10_scorers)

print("\n10 player terlama durasi bermain:")
print(top_10_duration)

print("\n10 player terbanyak assist:")
print(top_10_assists)

print("\n4 player pencetak gol terbanyak dari kesebelasan Manchester City:")
print(goals_manchester_city)

print("\nDari kesebelasan mana pencetak gol terbanyak:")
print(most_goals_team)

print("\nJumlah kartu kuning yang diterima pembuat assist terbanyak:", yellow_cards_most_assist)
print("Jumlah kartu merah yang diterima pembuat assist terbanyak:", red_cards_most_assist)

print("\nJumlah assist dan golnya paling banyak dan dari kesebelasan mana:")
print("Player:", most_goals_assists_player['player_name'])
print("Team:", most_goals_assists_team)
print("Jumlah assist dan gol:", most_goals_assists)

print("\n5 kesebelasan yang menyumbangkan gol terbanyak:")
print(top_5_goal_contributors)

print("\nUrutkan pemain berdasarkan jumlah gol terbanyak tanpa mendapatkan kartu sama sekali:")
print(players_without_cards)
```
